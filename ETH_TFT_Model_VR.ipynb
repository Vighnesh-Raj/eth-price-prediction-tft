    import torch
    import torch.nn as nn
    from torch.utils.data import Dataset, DataLoader
    from sklearn.model_selection import train_test_split
    import pandas as pd

    # 1. Dataset and DataLoader
    class CryptoDataset(Dataset):
        def __init__(self, data, input_columns, target_column, sequence_length):
            """
            Custom PyTorch Dataset for sequential data.

            Args:
            - data: DataFrame containing the data.
            - input_columns: List of input features.
            - target_column: Column to predict.
            - sequence_length: Number of time steps in the sequence.
            """
            self.data = data
            self.input_columns = input_columns
            self.target_column = target_column
            self.sequence_length = sequence_length

        def __len__(self):
            # Number of sequences that can be formed
            return len(self.data) - self.sequence_length

        def __getitem__(self, idx):
            # Extract sequence of inputs and the corresponding target
            x = self.data.iloc[idx:idx + self.sequence_length][self.input_columns].values
            y = self.data.iloc[idx + self.sequence_length][self.target_column]
            return torch.tensor(x, dtype=torch.float32), torch.tensor(y, dtype=torch.float32)

    # Define input features and target
    input_columns = [
        'ETH_Weekly_Moving_Average', 'Price_BTC', 'Open_BTC', 'High_BTC',
        'Low_BTC', 'Vol_BTC', 'Change_BTC', 'BTC_Weekly_Moving_Average', 'ETH_RSI'
    ]
    target_column = 'Change_ETH'
    sequence_length = 7  # 7-day sequence length

    # Train-test split
    train, temp = train_test_split(scaled_data, test_size=0.3, shuffle=False)
    val, test = train_test_split(temp, test_size=0.5, shuffle=False)

    # Create datasets
    train_dataset = CryptoDataset(train, input_columns, target_column, sequence_length)
    val_dataset = CryptoDataset(val, input_columns, target_column, sequence_length)
    test_dataset = CryptoDataset(test, input_columns, target_column, sequence_length)

    # Create data loaders
    batch_size = 32
    train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
    val_loader = DataLoader(val_dataset, batch_size=batch_size)
    test_loader = DataLoader(test_dataset, batch_size=batch_size)

    # 2. Temporal Fusion Transformer Model
    class TemporalFusionTransformer(nn.Module):
        def __init__(self, input_size, hidden_size, num_heads, dropout=0.1):
            """
            Temporal Fusion Transformer for time-series forecasting.

            Args:
            - input_size: Number of input features.
            - hidden_size: Hidden size of the transformer.
            - num_heads: Number of attention heads.
            - dropout: Dropout probability.
            """
            super(TemporalFusionTransformer, self).__init__()
            self.embedding = nn.Linear(input_size, hidden_size)
            self.transformer = nn.TransformerEncoder(
                nn.TransformerEncoderLayer(
                    d_model=hidden_size,
                    nhead=num_heads,
                    dropout=dropout,
                    activation='relu'
                ),
                num_layers=2
            )
            self.fc1 = nn.Linear(hidden_size, 64)
            self.fc2 = nn.Linear(64, 1)  # Output layer
            self.dropout = nn.Dropout(dropout)

        def forward(self, x):
            x = self.embedding(x)  # Shape: (batch_size, sequence_length, hidden_size)
            x = x.permute(1, 0, 2)  # Transformer expects (sequence_length, batch_size, hidden_size)
            x = self.transformer(x)
            x = x[-1, :, :]  # Keep the last output from the sequence
            x = torch.relu(self.fc1(x))
            x = self.dropout(x)
            x = self.fc2(x)
            return x

    # Model Initialization
    input_size = len(input_columns)
    hidden_size = 512
    num_heads = 8
    dropout = 0.2

    model = TemporalFusionTransformer(input_size=input_size, hidden_size=hidden_size, num_heads=num_heads, dropout=dropout)
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    model = model.to(device)

    # 3. Training the Model
    criterion = nn.MSELoss()
    optimizer = torch.optim.Adam(model.parameters(), lr=0.001)

    num_epochs = 20

    for epoch in range(num_epochs):
        model.train()
        train_loss = 0.0
        for inputs, targets in train_loader:
            inputs, targets = inputs.to(device), targets.to(device)

            # Forward pass
            outputs = model(inputs)
            loss = criterion(outputs.squeeze(), targets)

            # Backward pass and optimization
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()

            train_loss += loss.item()

        # Validation
        model.eval()
        val_loss = 0.0
        with torch.no_grad():
            for inputs, targets in val_loader:
                inputs, targets = inputs.to(device), targets.to(device)
                outputs = model(inputs)
                loss = criterion(outputs.squeeze(), targets)
                val_loss += loss.item()

        print(f"Epoch {epoch+1}/{num_epochs}, Train Loss: {train_loss/len(train_loader):.4f}, Val Loss: {val_loss/len(val_loader):.4f}")

    # Save the model
    torch.save(model.state_dict(), "tft_model.pth")

    # 4. Validation: Evaluate the model on test data
    model.eval()
    test_predictions = []
    test_actuals = []

    with torch.no_grad():
        for inputs, targets in test_loader:
            inputs, targets = inputs.to(device), targets.to(device)
            outputs = model(inputs)
            test_predictions.extend(outputs.cpu().numpy().flatten())
            test_actuals.extend(targets.cpu().numpy().flatten())

    # Results
    test_results = pd.DataFrame({
        'Actual': test_actuals,
        'Predicted': test_predictions
    })
    print(test_results.head())
